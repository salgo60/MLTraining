```python
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
data = pd.read_csv("StudentsPerformance.csv")
```

```python
data.shape, data.columns, data.info
```

```python
data.nunique()
```

```python
data["race/ethnicity"].unique(),data["parental level of education"].unique()
```

```python
data.isnull().sum()
```

```python
student = data.drop(['race/ethnicity','parental level of education'], axis=1)
```

```python
student
```

```python
correlation = student.corr()
```

```python
sns.heatmap(correlation,xticklabels=correlation.columns,yticklabels=correlation.columns)
```

```python
sns.pairplot(student)
```

```python
sns.relplot(x="math score",y="writing score",hue="gender",data=student)
```

```python
sns.distplot(student["writing score"],bins=5)
```

```python
sns.catplot(x="writing score",kind="box",data=student)
```

```python
student["writing score"].describe()
```

```python
sns.catplot(x="writing score",kind="violin",data=student)
```
